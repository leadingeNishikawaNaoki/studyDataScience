## 集合
集合Sに属していることを数式で表すと
$$
    x \in S
$$

集合の中身を具体的に書く場合は{}を使う
$$
    A = \{1, 2, 3, 4\}\\
    B = \{2, 4, 6, 8\}
$$

8はBに含まれるがAに含まれない表現
$$
    8 \in B\\
    8 \notin A
$$

2つの集合の共通部分を表す表現 = 積集合
$$
    A \cap B = \{2, 4\}
$$

2つの集合を全てまとめたもの = 和集合
$$
    A \cup B = \{1, 2, 3, 4, 6, 8\}
$$

空の集合 = 空集合
$$
    \emptyset
$$

# 数学記号
繰り返しを表現

## 足し算の繰り返し
ループの最初を下に、最後を上に書く。これは省略される場合もある。

x1からxnまでを全て足す計算の表現式
$$
    \sum_{i=1}^{n}x_i \tag{3.8}
$$

無限大まで足す
$$
    \sum_{n=1}^{\infty}\frac{1}{4^{n}} = \frac{1}{3} \tag{3.9}
$$

## 掛け算の繰り返し

x1からxnまでを全て掛け合わせた数の式の表現
$$
    \Pi_{i=1}^{n}x_i \tag{3.10}
$$

## 特殊な定数
eで表現される定数 ネイピア数(自然対数の底)
関数の微分や積分を扱う解析学の分野で重要
$$
    e = \sum_{n=0}^{\infty}\frac{1}{n!} \tag{3.11}
$$

n!は階乗でありnから順番に1まで掛け算した数
6! = 6x5x4x3x2x1=720
0! = 1

# 関数の基本
## 関数の書き方
次の関数fは、xを引数として受け取り、引数を２条して1を足した数を返す
$$
    f(x) = x^2 + 1 \tag{3.12}
$$

関数が複数の引数を持つ場合
$$
    f(x, y) = x^2 - y^2 + 2 \tag{3.13}
$$

## 少し特殊な関数
関数の定義が複数の数式からなることもある。

xが0から1の値を取る時には1を返し、そうでない場合は0を返す
$$
    f(x) = 
        \begin{cases}
            0, x > 1\\
            1, 0 \leq x \leq 1\\
            0, x < 0
        \end{cases}
    \tag{3.14}
$$

式(3.14)は式(3.15)のように書くことも可能
$$
    f(x) =
        \begin{cases}
            1, 0 \leq x \leq 1\\
            0, その他
        \end{cases}
    \tag{3.15}
$$
その他の部分は「otherwise」などの英語が使われたりもする

## 指数関数
$$
    f(x)=2^x
$$
という形で関数の入力が別の数字の肩に乗って使われる関数を、指数関数と呼ぶ
- この例では2を何乗するか？という意味
- この関数では、2は底(てい)と呼ばれる
- 底が1より大きな数の時、xが大きくなると、関数の値は大きくなる
- 底は整数でなくてもいい
- ネイピア数を底とする関数がよく用いられる
- 指数関数を応用した関数にシグモイド関数がある
- シグモイド関数はニューラルネットワークでよく使われる

$$
        f(x) = \frac{1}{1+e^-x} \tag{3.16}
$$

## 対数関数
$$
    f(x) = \log_2 x \tag{3.17}\\
    2^3 = 8 \Leftrightarrow 3 = \log_28
$$
この例ではf(8)=3
状況例では2を何乗すれば８になるか

$$
    f(x) = \log_e x = \ln x \tag{3.18}
$$

底が10の時は底が省略されることもある

## 三角関数
$$
    \tan \theta = \frac{\sin \theta}{\cos \theta} \tag{3.19}
$$

## 双曲線関数
- ハイパボリックサイン
- ハイパボリックコサイン
$$
    sinh x = \frac{e^x - e^-x}{2} \\
    cosh x = \frac{e^x + e^-x}{2} \tag{3.22}
$$

$$
    tanh x = \frac{sinh x}{cosh x} \tag{3.23}
$$

# 線形代数
- ベクトルとは丸括弧で数をまとめて表現したものである
- 原点を始点とするベクトルを、位置ベクトルと呼ぶ

$$
    (4, 7) \tag{3.24}
$$
式(3.24)は二次元ベクトル

- 場所を意味する座標を表現する場合は、数字を横に並べて書き、ベクトルを表現する場合は縦に並べて書く方法もある
$$
    \left( \begin{array}{c} 4 \\ 7 \end{array} \right) \tag{3.25}
$$

- ベクトルとの対比で異なる1つの数をスカラーと呼ぶ
- ベクトルとスカラーは掛け算ができる

## ノルム
- ベクトルの大きさをスカラーで表現する時、ノルムと呼ぶ
- ベクトルを表面上の原点からの直線距離として測り表現する = ベクトルの両脇にパイプを2つ書く

$$
    \begin{vmatrix}
    x
    \end{vmatrix}
    = \sqrt{x^2_1 + x^2_2 + \cdots + x^2_n} \tag{3.30}
$$
